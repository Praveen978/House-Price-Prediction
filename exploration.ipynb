# House Prices - Exploratory Data Analysis (EDA)

This notebook explores the house_prices.csv dataset.

```python
import pandas as pd

df = pd.read_csv('../data/house_prices.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
df.info()
df.describe()
```
